```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# import database
df = pd.read_csv('User_Datetime_Analysis.csv')
df
```

```python
# to show current time
current_time = datetime.now() #timestamp date
print(current_time)
```

```python
df['Signup_Date']
```

```python
# To show the data in datetime format it needs to be converted from an object to datetime via assingment
df['Signup_Date'] = pd.to_datetime(df['Signup_Date'])
```

```python
df['Signup_Date']
```

```python
df['Last_Login']
```

```python
df['Last_Login'] = pd.to_datetime(df['Last_Login'])
df['Last_Login']
```

```python
# workout sign up date (month)

df['Signup_Month'] = df['Signup_Date'].dt.month
df['Signup_Month']
```

```python
# workout differences in dates (days since last login)
df['Days_Since_Last_Login'] = (pd.Timestamp.now() - df['Last_Login']).dt.days
df['Days_Since_Last_Login']
```

```python
# workout sign up date (month name)

df['Signup_Month_Name'] = df['Signup_Date'].dt.month_name()
df['Signup_Month_Name']
```

```python
# workout sign up date (day name)

df['Signup_Day_Name'] = df['Signup_Date'].dt.day_name()
df['Signup_Day_Name']
```

```python
# workout oldest date entry in a column (oldest person in database)

oldest = df.sort_values('Birthdate').head(1)
oldest
```

```python
# workout youngest date entry in a column (youngest person in database)
# head() shows first, tail() shows last

youngest = df.sort_values('Birthdate').tail(1)
youngest
```

```python
# convert to datetime for hours minutes and seconds

df['Last_Login'] = pd.to_datetime(df['Last_Login'], format = '%A, %d %B %Y %H:%M')
# time_now = datetime.striptime('2025-05-15 19:11:09.021056', '%Y-%m-%d %H %m %s')
```

```python
df['Last_Login']
```

```python
df['Hour'] = df['Last_Login'].dt.hour
df['Hour']
```

```python
df['Minute'] = df['Last_Login'].dt.minute
df['Minute']
```

```python
df['Second'] = df['Last_Login'].dt.second
df['Second']
```

```python
# Filtering
# workout inactivity within the last 20 days
twenty_day_inactivity = df[df['Days_Since_Last_Login'] > 20]
twenty_day_inactivity
```

```python
df['Signup_Month_Name'] = df['Signup_Date'].dt.month_name()
df['Signup_Month_Name']
```

```python
sign_up_nov = df[df['Signup_Month_Name'] == 'November']
sign_up_nov
```
