```python
# import necessary libraries
import pandas as pd

# import database
df = pd.read_csv("Updated_Drug_Dataset_with_Real_Dates.csv")
df
```

```python
# Number of country per price usd (save this in csv)

# grouby function - showing sum price_usd per country
country_sum = df.groupby('Country')['Price_USD'].sum()
country_sum
```

```python
# Save filtered data as a new file without Index
df.to_csv("Filtered_File_Without_Index.csv", index = False)
```

```python
# Number Is_generic per dosage_Mg (save this excel)

# grouby function - showing if dosage is generic
dosage_is_generic = df.groupby('Dosage_mg')['Is_Generic'].sum()
dosage_is_generic
```

```python
# Save filtered data as a new file without Index
df.to_excel("Filtered_File_Without_Index.xlsx", index = False)
```

```python
# Rename the Drug_id to Medication_ID
df.rename(columns = {'Drug_Name': 'Medication_Name', 'Drug_ID': 'Medication_ID'}, inplace = True)
df
```

```python
#Add the name of the organization to the dataframe "Adech Global"
df['Organisation_Name'] = 'Adech Global'
df
```

```python
#Country, Germany:01, India:09,  UK:04, USA: 07, Canada: 06
Country_map = {'Germany': '01', 'India': '09', 'UK': '04', 'USA': '07', 'Canada': '06'}
df['Country'] = df['Country'].map(Country_map)
df
```

```python
#Fill all the missing numbers in the data with unknown

df['Medication_ID'] = df['Medication_ID'].fillna(value = 'Missing')
df['Medication_Name'] = df['Medication_Name'].fillna(value = 'Missing')
df['Manufacturer'] = df['Manufacturer'].fillna(value = 'Missing')
df['Dosage_mg'] = df['Dosage_mg'].fillna(value = 'Missing')
df['Form'] = df['Form'].fillna(value = 'Missing')
df['Approval_Year'] = df['Approval_Year'].fillna(value = 'Missing')
df['Price_USD'] = df['Price_USD'].fillna(value = 'Missing')
df['Side_Effects'] = df['Side_Effects'].fillna(value = 'Missing')
df['Is_Generic'] = df['Is_Generic'].fillna(value = 'Missing')
df['Country'] = df['Country'].fillna(value = 'Missing')
df['Organisation_Name'] = df['Organisation_Name'].fillna(value = 'Missing')

# for i in df:
#     if i == 'Nan':
#       df[i] = df[i].fillna(value = 'Missing')  
df
```
