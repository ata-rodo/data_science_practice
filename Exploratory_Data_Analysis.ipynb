```python
# Week 8 - # EDA (Exploratory Data Analysis)

# import neccesary libraries
import pandas as pd
import numpy as np

# import data
df = pd.read_csv("Updated_Drug_Dataset_with_Real_Dates.csv")
df
```

```python
# to isolate duplicated rows

duplicated_rows = df[df.duplicated]
duplicated_rows
```

```python
# drop duplicated rows
drop_duplicates = df.drop_duplicates()
drop_duplicates

# drop duplicated rows AND view table
drop_duplicates = df[df.duplicated()]
drop_duplicates
```

```python
# drop duplicates in specific column
drop_form_duplicates = df[df.duplicated('Form', keep = False)]
drop_form_duplicates
```

```python
# groupby function
manufacturer_count = df.groupby('Manufacturer')['Dosage_mg'].sum()
manufacturer_count
```

```python
# grouby function - showing sum of each drug
drug_sum = df.groupby('Drug_Name')['Price_USD'].sum()
drug_sum

# len() function shows total number of categories
len(drug_sum)
```

```python
# Unique function - shows different categories in a column and highlight empty/missing data
df['Form'].unique()
```

```python
df['Manufacturer'].unique()
```

```python
# rename a column in dataframe
df.rename(columns = {'Drug_Name': 'Medication_Name'}, inplace = True)
df
```

```python
df['Branch_No.'] = 106
df
```

```python
# GDPR General Data Protection Regulation
# to anomise (make anonymous) sensitive data

df['Manufacturer'].unique()
```

```python
# CureAll: 01
# Wellness Inc.5: 02
# MediLife: 03
# PharmaCorp: 04
# HealthPlus: 05

# use mapping to make data anonymous

Manufacturer_map = {'CureAll': '01', 'Wellness Inc.': '02', 'MediLife': '03', 'PharmaCorp': '04', 'HealthPlus': '05'}
df['Manufacturer'] = df['Manufacturer'].map(Manufacturer_map)
# df
```
